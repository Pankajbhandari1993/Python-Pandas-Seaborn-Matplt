```python
numeric_cols = [
    "Industrial_Output_Crore",
    "Employment_Lakh",
    "Factories_Count",
    "Pollution_Index",
    "Electricity_Consumption_GWh",
    "Export_Crore"
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# STEP 12   GroupBy Analysis
```

```python
state_output = df.groupby("State")["Industrial_Output_Crore"].sum().astype(int)
```

```python
industry_emp = df.groupby("Industry_Type")["Employment_Lakh"].mean().astype(float)
```

```python
year_investment = df.groupby("Year")["Investment_Crore"].sum().astype(int)
```

```python
# Visualization
```

```python
# Bar chart showing Industrial Output Share of Top 5 States


top5 = state_output.sort_values(ascending=False).head(5)  # State_output is groupby variable name calculated above

plt.figure(figsize=(8,8))

plt.pie(
    top5,
    labels=top5.index,
    autopct="%1.1f%%"
)

plt.title("Industrial Output Share of Top 5 States")

plt.show()
```

```python
# Bar Chart showing Average Employment by Industry

plt.figure(figsize=(12,6))

ax=industry_emp.sort_values().plot(kind="bar")  # industry_emp is a group by calculated above

plt.title("Average Employment by Industry")

plt.xlabel("Industry")

plt.ylabel("Employment")
for container in ax.containers:             # For data labels
    ax.bar_label(container, fmt='%.1f', padding=3)

plt.xticks(rotation=45)

plt.show()
```
